```python
#BMI calc
name1 = "TP"
height_m = 175
weight_kg1 = 80

name2 = "BK"
height_m2 = 173
weight_kg2 = 90

name3 = "Tomi"
height_m3 = 180
weight_kg3 = 100

def bmi_calculator (name, height_m, weight_kg):
    bmi = weight_kg / (height_m ** 2)
    print ("bmi")
    print (bmi)
    if bmi < 25:
        return name + " is not overweight"
    else:
        return name + " is overwight"
result1 = bmi_calculator (name1,  height_m,    weight_kg1)
result2 = bmi_calculator (name2,  height_m2,  weight_kg2)
result3 = bmi_calculator (name3,  height_m3,  weight_kg3)

print(result1)
print(result2)
print(result3)
 
```
